```python
#Fuzzy Control Systems: Learning a new language
```

```python
#This fuzzy control system is designed to suggest a practice level when learning a new language based on a user's proficiency 
#level and the difficulty of the learning content. It involves three main components:

#1. Fuzzy Variables (Antecedents and Consequent):
#Proficiency Level: Represents the learner's language proficiency, defined with three fuzzy sets: 'poor', 'average', and 'good'.
#Content Difficulty: Represents how difficult the content is, with three fuzzy sets: 'easy', 'average', and 'hard'.
#Recommended Practice: The output that suggests the level of practice the learner should undertake. It has three 
#fuzzy sets: 'basic', 'standard', and 'advanced'.

#2. Fuzzy Rules:
#The system uses three fuzzy logic rules to map the relationship between proficiency level, content difficulty, and recommended 
#practice:
#Rule 1: If the proficiency level is 'poor' OR content difficulty is 'hard', then recommend 'basic' practice.
#Rule 2: If the proficiency level is 'average', recommend 'standard' practice.
#Rule 3: If the proficiency level is 'good' OR content difficulty is 'easy', recommend 'advanced' practice.

#3. Inference and Defuzzification:
#The fuzzy rules combine inputs and generate a fuzzy output, which is then converted into a crisp value using defuzzification 
#methods. In this case, five defuzzification methods (centroid, bisector, mean of maximum, smallest of maximum, largest of maximum)
#are applied, and their results are visualized to show how each method affects the output recommendation.

#4. Graphical Visualization:
#The membership functions for each variable are visualized, showing how values fall within the fuzzy sets. The defuzzification 
#process is also plotted, where different methods outputs are compared against the fuzzy membership function to see the 
#recommended practice level.
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# The Antecedent and Consequent objects
proficiency_level = ctrl.Antecedent(np.arange(0, 11, 1), 'proficiency_level')
content_difficulty = ctrl.Antecedent(np.arange(0, 11, 1), 'content_difficulty')
recommended_practice = ctrl.Consequent(np.arange(0, 11, 1), 'recommended_practice')

# Auto-membership function population for proficiency_level
proficiency_level.automf(3)  # Creates automatic fuzzy sets: 'poor', 'average', 'good'
```

```python
# Custom membership functions for content difficulty
content_difficulty['easy'] = fuzz.trimf(content_difficulty.universe, [0, 0, 5])
content_difficulty['average'] = fuzz.trimf(content_difficulty.universe, [0, 5, 10])
content_difficulty['hard'] = fuzz.trimf(content_difficulty.universe, [5, 10, 10])

# Custom membership functions for recommended practice
recommended_practice['basic'] = fuzz.trimf(recommended_practice.universe, [0, 0, 5])
recommended_practice['standard'] = fuzz.trimf(recommended_practice.universe, [0, 5, 10])
recommended_practice['advanced'] = fuzz.trimf(recommended_practice.universe, [5, 10, 10])

# View the membership functions
proficiency_level.view()
content_difficulty.view()
recommended_practice.view()
```

```python
# Fuzzy Rules
rule1 = ctrl.Rule(proficiency_level['poor'] | content_difficulty['hard'], recommended_practice['basic'])
rule2 = ctrl.Rule(proficiency_level['average'], recommended_practice['standard'])
rule3 = ctrl.Rule(proficiency_level['good'] | content_difficulty['easy'], recommended_practice['advanced'])

#Rule 1: If proficiency level is poor OR content difficulty is hard, then recommended practice is basic.
#Rule 2: If proficiency level is average, then recommended practice is standard.
#Rule 3: If proficiency level is good OR content difficulty is easy, then recommended practice is advanced.
```

```python
# Create the control system
language_learning_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])

# Create a ControlSystemSimulation object
learning_system = ctrl.ControlSystemSimulation(language_learning_ctrl)
```

```python
# Let's say a learner has a proficiency level of 4 (average) and the content is of difficulty 7 (hard)
learning_system.input['proficiency_level'] = 4
learning_system.input['content_difficulty'] = 7

# Do the calculations
learning_system.compute()

# Display the recommended practice level
print(f'Recommended practice level: {learning_system.output["recommended_practice"]}')
recommended_practice.view(sim=learning_system)
```

```python
import matplotlib.pyplot as plt

# Generate a trapezoidal membership function for visualization
x = np.arange(0, 11, 0.1)
mfx = fuzz.trapmf(x, [2, 3, 5, 7]) 

# Defuzzify this membership function using different methods
defuzz_centroid = fuzz.defuzz(x, mfx, 'centroid') 
defuzz_bisector = fuzz.defuzz(x, mfx, 'bisector')
defuzz_mom = fuzz.defuzz(x, mfx, 'mom')
defuzz_som = fuzz.defuzz(x, mfx, 'som')
defuzz_lom = fuzz.defuzz(x, mfx, 'lom')

# Print the defuzzification values
print(f"Defuzzification values:")
print(f"Centroid: {defuzz_centroid}")
print(f"Bisector: {defuzz_bisector}")
print(f"Mean of Maximum: {defuzz_mom}")
print(f"Min of Maximum: {defuzz_som}")
print(f"Max of Maximum: {defuzz_lom}")

# Collect information for vertical lines in the plot
labels = ['centroid', 'bisector', 'mean of maximum', 'min of maximum', 'max of maximum']
xvals = [defuzz_centroid, defuzz_bisector, defuzz_mom, defuzz_som, defuzz_lom]
colors = ['r', 'b', 'g', 'c', 'm']
ymax = [fuzz.interp_membership(x, mfx, i) for i in xvals]

# Display and compare defuzzification results against the membership function
plt.figure(figsize=(8, 5))

plt.plot(x, mfx, 'k')  # Plot the trapezoidal membership function
for xv, y, label, color in zip(xvals, ymax, labels, colors):
    plt.vlines(xv, 0, y, label=label, color=color)  # Draw vertical lines at defuzzification points

# Format the plot
plt.ylabel('Fuzzy membership')
plt.xlabel('Recommended practice level')
plt.ylim(-0.1, 1.1)
plt.legend(loc=2)

plt.show()
```
